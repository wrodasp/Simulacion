```python
from datetime import datetime
from scipy.optimize import curve_fit
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline

#Lectura de datos y filtrado para Colombia
url = 'http://cowid.netlify.com/data/full_data.csv'

datos = pd.read_csv(url)

datos = datos[datos['location'].isin(['Russia'])] #Filtro la Informacion solo para Ecuador
datos = datos.loc[:,['date','total_cases']] #Selecciono las columnas de analasis

# Expresar las fechas en numero de dias desde el 01 Enero
formato = '%Y-%m-%d'
fecha = datos['date']
datos['date'] = fecha.map(lambda x : (datetime.strptime(x, formato) - datetime.strptime("2020-01-01", formato)).days)

#Obtención de los datos para las variables días y total_casos
dias = list(datos.iloc [:, 0])
total_casos = list(datos.iloc [:, 1])

#Obtención del día objetivo
dia_objetivo = dias[-1] + 7

datos
```

```python
datos.plot(x ='date', y='total_cases')
```

### Lineal

```python
modelo_lineal = LinearRegression()
modelo_lineal.fit(np.array(dias).reshape(-1, 1), total_casos)

prediccion_semana = modelo_lineal.predict([[dia_objetivo]])
print('Predicción:', int(prediccion_semana))

rango_dias = np.arange(60, 85, 1)

plt.scatter(dias, total_casos , color = 'red')
plt.plot(rango_dias, modelo_lineal.predict(rango_dias.reshape(-1, 1)), color = 'green')
plt.show()
```

### Logistico

```python
modelo_logistico = lambda x, a, b: a + b * np.log(x)

ajuste = curve_fit(modelo_logistico, dias, total_casos)

rango_dias = np.arange(60, 85, 1)

prediccion_total = [modelo_logistico(i, ajuste[0][0], ajuste[0][1]) for i in rango_dias]

prediccion_semana = prediccion_total[-3]
print('Predicción: ', int(prediccion_semana))

plt.scatter(dias, total_casos , color = 'red')
plt.plot(rango_dias, prediccion_total, color = 'green')
plt.show()
```

### Exponencial

### Polinomial 

```python
definicion = PolynomialFeatures(degree = 5)
modelo_polinomial = LinearRegression()

Dias = definicion.fit_transform(np.array(dias).reshape(-1, 1))

modelo_polinomial.fit(Dias, total_casos)

prediccion_semana = modelo_polinomial.predict([Dias[-1] + 7])
print('Predicción: ', int(prediccion_semana))

rango_dias = np.array(range(60, 85))
rango_dias = definicion.fit_transform(rango_dias.reshape(-1, 1))

plt.scatter(dias, total_casos, color = 'red')
plt.plot(modelo_polinomial.predict(rango_dias), color = 'green')
plt.show()
```

### Probabilístico

```python
mediana = datos['total_cases'].median()

modelo_probabilistico = lambda y: int(y[-1] + mediana)

prediccion = total_casos

for i in range(dias[-1], dia_objetivo):
    dias.append(i)
    prediccion.append(modelo_probabilistico(prediccion))
    
prediccion_semana = modelo_probabilistico(total_casos)
print('Predicción: ', int(prediccion_semana))

plt.scatter(dias, total_casos, color = 'red')
plt.plot(dias, total_casos, color = 'green')
plt.show()
```
