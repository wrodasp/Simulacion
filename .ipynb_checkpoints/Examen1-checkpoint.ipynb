### 3. Buscando los tweets de los candidatos

```python
#Definimos una lista donde se guardaran los datos
datos = []

#Realizamos la consulta y obtenemos los resultados
for usuario in USUARIOS:
    tweets = twitter.user_timeline(usuario, tweet_mode = 'extended')

    #Guardamos los datos en un diccionario
    for tweet in tweets:
        datos.append({
            'candidato': tweet.user.name,
            'fecha': tweet.created_at,
            'retweets': tweet.retweet_count,
            'favoritos': tweet.favorite_count,
            'tweet': tweet.full_text
    })

#Convertimos los datos en dataframe para facilitar el trabajo
datos = pd.DataFrame(data = datos, columns = ['candidato', 'fecha', 'retweets', 'favoritos','tweet'])

#Mostramos los 10 primero datos obtenidos
datos
```

### 4. Preprocesando los datos

```python
#Expresar los datos de fecha en número de dias desde inicio del año 2020
formato = '%Y-%m-%d'
fecha = datos['fecha']

datos['fecha'] = fecha.map(lambda x: 
    (datetime.strptime(str(x.date()), formato) - datetime.strptime('2020-01-01', formato)).days
)

datos
```

### 5. Gráfico inicial (Tendencia de apoyo)

### 6.1. Predicción de tendencia dentro de 40 dias, mediante regresión polinomial

```python
grafico = pp.figure(figsize = (15, 40))

i = 1
dias_totales = np.array(range(320, 400))

for etiqueta, grupo in partidos:
    #Obtención de las variables para el entrenamiento
    x = list(grupo.iloc[:, 1]) #Fecha (Número de día)
    y = list(grupo.iloc[:, 2]) #Retweets obtenidos

    #Creación y entranamiento del modelo
    
    modelo = LinearRegression()
    modelo.fit(np.array(x).reshape(-1, 1), y)

    #Establecemos el número de días que queremos predecir
    DIAS_PREDICCION = x[-1] + 40

    #Realizamos la prediccion
    prediccion = modelo.predict([[DIAS_PREDICCION]])
    prediccion = max(x) if int(max(x) + prediccion) <= 0 else int(max(x) + prediccion)
    
    candidato = grupo.iloc[0,0]
    
    parte = grafico.add_subplot(10, 2, i)
    parte.set_title(candidato + ' - Total de retweets: ' + str(prediccion))
    parte.scatter(x, y, color = 'green')
    parte.plot(dias_totales, modelo.predict(dias_totales.reshape(-1, 1)), color = 'red')
    parte.legend(['Retweets actuales', 'Retweets predecidos'])

    i += 1
    
pp.show()
```

### 1. Definición de datos iniciales

A continuación se establecen los parámetros de la simulación, suponiendo un ambiente ficticio donde ocurre lo siguiente:

2. Debido al COVID-19 todos los recintos electorales del Azuay, a excepción de la Universidad de Cuenca, sufren cierres para desinfección y el Ministerio de Salud prohíbe usarlas.

3. Al ser un evento demasiado tardado cada persona se demora entre 2 y 3 minutos como máximo para integrarse en la cola. 

4. Cada persona se demora entre 1 y 2 minutos para ejercer su voto.

5. A este proceso asisten al menos el 90% de los habitantes del Azuay.

8. El periodo de sufragio comienza el 07/02/2021 a las 6:00 am y partir de ahí podría durar entre 12 y 16 horas para atender a la mayoría de los Azuayos.

### 2. Creación del modelo y sus recursos

```python
class Recinto(object):
    
    def __init__(self, entorno, carton_votos, cantidad_biombos, 
                 tiempo_votacion_minimo, tiempo_votacion_maximo,
                 registro_tiempos_votacion, registro_tiempos_llegada):
        self.entorno = entorno
        self.carton_votos = carton_votos
        self.biombos = sp.Resource(entorno, cantidad_biombos)
        self.tiempo_votacion_minimo = tiempo_votacion_minimo
        self.tiempo_votacion_maximo = tiempo_votacion_maximo
        self.registro_tiempos_votacion = registro_tiempos_votacion
        self.registro_tiempos_llegada = registro_tiempos_llegada
        self.fecha = datetime(2021, 2, 7, 7, 0, 0)
        self.biombos_libres = cantidad_biombos
        
    def permitir_sufragar(self, persona, total_nulos):
        #La persona realiza el sufragio entre 1 y 2 minutos
        tiempo_votacion = rd.randint(self.tiempo_votacion_minimo, self.tiempo_votacion_maximo)
        tiempo_consumido = self.fecha + timedelta(minutes = tiempo_votacion)
        
        #Se realiza el voto por un candidato o nulo
        if total_nulos == 0:
            self.carton_votos[rd.randint(0, 16)][1] += 1
        else:
            self.carton_votos[17][1] += 1
            
        self.registro_tiempos_votacion.append(tiempo_votacion)
        imprimir_registro('El votante #{} sufragó | {}'.format(persona, tiempo_consumido))
        
        yield self.entorno.timeout(tiempo_votacion)
        
        #La persona recibe su certificado de votación
        self.biombos_libres += 1
        imprimir_registro('El votante #{} recibió su certificado de votación.'.format(persona))
        
        #La persona se retira del recinto
        tiempo_consumido = self.fecha + timedelta(seconds = rd.randint(0, 60))
        imprimir_registro('El votante #{} salió del recinto | {}.'.format(persona, tiempo_consumido))

def llegada_persona(entorno, recinto, persona, intervalo, total_nulos):        
    #Muestra como llegan los sufragantes
    tiempo_llegada = recinto.fecha + timedelta(minutes = intervalo)
    recinto.registro_tiempos_llegada.append(intervalo)
    imprimir_registro('Llega el votante #{} | {}'.format(persona, tiempo_llegada))
    
    #Las personas usan el recinto electoral
    with recinto.biombos.request() as biombo:
        #Ocupamos el biombo solicitado
        recinto.biombos_libres -= 1
        yield biombo
        
        #Realizamos el proceso de sufragación
        yield entorno.process(recinto.permitir_sufragar(persona, total_nulos))
        
def imprimir_registro(mensaje):
    global contador_registros
    global registros_a_omitir
    global registros_a_imprimir
    
    if contador_registros < MAXIMO_REGISTROS:
        if registros_a_omitir > 0:
            registros_a_omitir -= 1
        else:
            if registros_a_imprimir < int(total_personas / MAXIMO_REGISTROS):
                print(mensaje)
                contador_registros += 1
                registros_a_imprimir += 1
            else:
                registros_a_omitir = MAXIMO_REGISTROS
                registros_a_imprimir = 0
            
            
        
def calcular_total_nulos(total_personas):
    #Se calcula el total de personas del grupo que votarán nulo
    porcentaje_nulos_minimo = PORCENTAJE_NULOS_MINIMO * 100
    porcentaje_nulos_maximo = PORCENTAJE_NULOS_MAXIMO * 100

    porcentaje_nulos = rd.randint(porcentaje_nulos_minimo, porcentaje_nulos_maximo) / 100
    
    return int(total_personas * porcentaje_nulos)
        
def iniciar_sufragio(entorno, recinto, total_personas):
    print('El proceso electoral de asambleistas ha comenzado!\n')
    
    intervalo_promedio = 0
    #Iniciamos el proceso electoral con los miembros de las juntas receptoras del voto
    #
    #Esto se hace debido a que generalmente los miebros de mesa
    #sufragan primero, con el fin de evitar inconvenientes al 
    #resto de personas.
    miembros_mesa = CANTIDAD_MESAS * MIEMBROS_POR_MESA
    
    total_nulos = calcular_total_nulos(miembros_mesa)
    
    for persona_actual in range(miembros_mesa):
        entorno.process(
            llegada_persona(
                entorno,
                recinto,
                persona_actual + 1,
                intervalo_promedio, 
                total_nulos
            )
        )
        
        if total_nulos > 0: 
            total_nulos -= 1
        
    while True:
        #Las personas van llegando entre 2 y 3 minutos
        intervalo_promedio = rd.randint(INTERVALO_LLEGADA, INTERVALO_LLEGADA + 1)
        yield entorno.timeout(intervalo_promedio)
        
        #Se dejan pasar N cantidad de personas para usar lo biombos libres
        grupo_personas = recinto.biombos_libres
        
        total_nulos = calcular_total_nulos(grupo_personas)
        
        for persona in range(grupo_personas):
            persona_actual +=1
            
            #Se verifica que solo sufraguen el limite de personas del Azuay
            if persona_actual <= total_personas:
                entorno.process(
                    llegada_persona(
                        entorno,
                        recinto,
                        persona_actual,
                        intervalo_promedio,
                        total_nulos
                    )
                )
                
                if total_nulos > 0: 
                    total_nulos -= 1
        
        #Se suman 3 minutos a la hora del dia
        recinto.fecha = recinto.fecha + timedelta(minutes = 3)
```

### 3. Realizamos el proceso de simulación

```python
#Calculamos el total de habitantes que van a sufragar
total_personas = int(HABITANTES_AZUAY * PORCENTAJE_VOTANTES)

#Calculamos el total de votos nulos que habrá
carton_votos = []

for partido in PARTIDOS:
    carton_votos.append([partido, 0])

#Definimos variables utiles para limitar los registros a imprimir en consola
contador_registros = 0
registros_a_omitir = MAXIMO_REGISTROS
registros_a_imprimir = 0

#Definimos el tiempo de simulacion
#Unidades de tiempo simulado en minutos
tiempo_a_simular = (total_personas * INTERVALO_LLEGADA * TIEMPO_VOTACION_MAXIMO) / CANTIDAD_BIOMBOS 

#Creamos el entorno de simulación
entorno = sp.Environment()

#Creamos nuestro recinto
recinto = Recinto(
    entorno,
    carton_votos,
    CANTIDAD_BIOMBOS, 
    TIEMPO_VOTACION_MINIMO, 
    TIEMPO_VOTACION_MAXIMO,
    [], []
)

entorno.process(
    iniciar_sufragio(
        entorno,
        recinto,
        total_personas
    )
)

print('Total de personas a votar:', total_personas, '\n')

#Ejecutamos la simulación
inicio_simulacion = datetime.now()

entorno.run(until = tiempo_a_simular)

fin_simulacion = datetime.now()

print('\nEl proceso electoral ha finalizado!')
```

En la simulación realizada hay que considerar que las primeras N personas (Miembros de las juntas receptoras del voto), aunque se registra su salida, se asume que no se retiran del recinto electoral hasta cumplido con el día de sufragio. De acuerdo a los detalles presentados previamente para el establecimiento de parámetros, esta cantidad N = CANTIDAD_MESAS * MIEMBROS_POR_MESA.

### 4. Resultados de Tiempos

```python
#Obtenemos los registros de tiempos de votacion
registro_tiempos = np.array(recinto.registro_tiempos_votacion)
tiempo_promedio_votacion = np.sum(registro_tiempos) / len(registro_tiempos)

print('\nTiempo promedio de sufragación: {:.2f} minutos'.format(tiempo_promedio_votacion))

#Obtenemos los registros de tiempos de espera
registro_tiempos = np.array(recinto.registro_tiempos_llegada)
tiempo_promedio_espera = np.sum(registro_tiempos) / len(registro_tiempos)

print('\nTiempo promedio de espera en la cola: {:.2f} minutos'.format(tiempo_promedio_espera))

#Calculamos el tiempo simulado
tiempo_simulacion = tiempo_a_simular / 60

print('\nTiempo simulado: {:.0f} horas'.format(tiempo_simulacion))

#Calculamos el total de tiempo de simulacion en la computadora (Tiempo real)
tiempo_real = (fin_simulacion - inicio_simulacion).total_seconds() / 60

print('\nTiempo real de simulación: {:.2f} minutos'.format(tiempo_real))
```
